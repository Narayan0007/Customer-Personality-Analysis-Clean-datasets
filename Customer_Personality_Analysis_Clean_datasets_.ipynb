```python
# Import required Library
import pandas as pd
df = pd.read_csv("/content/marketing_campaign.csv", sep='\t')
df.head()
```

```python
# Showing all columns
pd.set_option("display.max_columns", None)
df.head()
```

```python
# Show all necessery information
df.info()
```

```python
# Data Shape and Summary Statistics
print("Data Shape:", df.shape)
df.describe(include='all')
```

```python
# Import required Library
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from datetime import datetime
```

```python
# Data Cleaning
# 1. Handle missing values
print("Missing values :")
print(df.isnull().sum())
```

```python
# Drop rows with missing income (only 24 out of 2240)
df = df.dropna(subset=['Income'])
```

```python
# Cheking duplicates rows
df.duplicated().sum()
```

```python
# No duplicate rows were found in the dataset.
```

```python
# Checking dtypes
print(df.dtypes)
```

```python
# Convert Dt_Customer to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
# Create age feature from Year_Birth
current_year = datetime.now().year
df['age'] = current_year - df['Year_Birth']
```

```python
df.head()
```

```python
# Rename column headers to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
# Create total children feature
df['total_children'] = df['kidhome'] + df['teenhome']
```

```python
# Create total spending feature
df['total_spending'] = df[['mntwines', 'mntfruits', 'mntmeatproducts',
                          'mntfishproducts', 'mntsweetproducts', 'mntgoldprods']].sum(axis=1)
```

```python
# Create total accepted campaigns
df['Total_Accepted_Cmp'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3',
                              'AcceptedCmp4', 'AcceptedCmp5', 'Response']].sum(axis=1)
```

```python
# Drop unnecessary columns
df = df.drop(['z_costcontact', 'z_revenue'], axis=1)
```

```python
# Simplifying marital status to Relationship/Single
df['marital_status'].unique()
```

```python
df['marital_status'] = df['marital_status'].replace(['Together', 'Married'] , 'Relationship' )
df['marital_status'] = df['marital_status'].replace(['Divorced', 'Widow', 'Alone', 'Absurd', 'YOLO'] , 'Single')
```

```python
df.head()
```

```python
df.to_csv('cleaned_marketing_campaign.csv', index=False)
print("DataFrame successfully saved to 'cleaned_marketing_campaign.csv'")
```
